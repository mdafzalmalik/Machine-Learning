```python
import pandas as pd

data = {
    "Year": [2008, 2009, 2010, 2011, 2012],
    "Rohit Sharma Runs": [500, 700, 650, 800, 750],
    "Kohli Runs": [400, 600, 700, 900, 850]
}

df = pd.DataFrame(data)

print(df)
```

```python
plt.plot(data["Year"], data["Rohit Sharma Runs"], color = "red")
plt.plot(data["Year"], data["Kohli Runs"], color = "green")
plt.title("Rohit Sharma and Kohli Runs comparision")
plt.xlabel("Seasons")
plt.ylabel("Runs")
plt.show()
```

```python
ypoints = data["Year"]
plt.plot(ypoints)
```

```python
plt.plot(data["Year"], data["Rohit Sharma Runs"], color = "red", linestyle="dashed", marker='d')
plt.plot(data["Year"], data["Kohli Runs"], color = "green", linewidth=3, marker='o')
plt.title("Rohit Sharma and Kohli Runs comparision")
plt.xlabel("Seasons")
plt.ylabel("Runs")
plt.show()
```

```python
plt.plot(data["Year"], data["Rohit Sharma Runs"], 'o-.b', label="Rohit")
plt.plot(data["Year"], data["Kohli Runs"], color = "green", linewidth=3, marker='o', label="Kohli")
plt.title("Rohit Sharma and Kohli Runs comparision")
plt.xlabel("Seasons")
plt.ylabel("Runs")
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(data["Year"], data["Rohit Sharma Runs"], color = "red", linestyle="dashed", marker='d')
plt.subplot(1,2,2)
plt.plot(data["Year"], data["Kohli Runs"], color = "green", linewidth=3, marker='o')
plt.suptitle("Rohit Sharma and Kohli Runs comparision")
plt.xlabel("Seasons")
plt.ylabel("Runs")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Year": [2008, 2009, 2010, 2011, 2012],
    "Rohit Sharma Runs": [500, 700, 650, 800, 750],
    "Kohli Runs": [400, 600, 700, 900, 850]
}

df = pd.DataFrame(data)

# Bar chart for Rohit Sharma
plt.bar(df["Year"], df["Rohit Sharma Runs"], color="skyblue", edgecolor="black")
plt.title("Rohit Sharma Runs per Year")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = [2008, 2009, 2010, 2011, 2012]
rohit_runs = [500, 700, 650, 800, 750]
kohli_runs = [400, 600, 700, 900, 850]

x = np.arange(len(years))  # positions for bars
width = 0.4  # width of each bar

# Simple bivariate bar graph
plt.bar(x - width/2, rohit_runs, width, label="Rohit Sharma", color="skyblue", edgecolor="black")
plt.bar(x + width/2, kohli_runs, width, label="Kohli", color="salmon", edgecolor="black")

plt.xticks(x, years)  # set x-axis labels
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Runs per Year: Rohit Sharma vs Kohli")
plt.legend()
plt.show()
```
