```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/titanic.csv")
```

```python
df.head()
```

```python
# generally we start from left to write but currently we are learning so we will first do eda for categorical then numerical
```

1. Categorical data

a. countplot

```python
sns.countplot(
    x = "Survived",
    data = df,
    hue = "Survived",
    palette = ["red", "green"],
)


# df["Survived"].value_counts().plot(kind = "bar")
```

```python
sns.countplot(
    x = "Pclass",
    data = df,
    hue = "Pclass"
)
```

```python
sns.countplot(
    x = "Sex",
    data = df,
    hue = "Sex"
)
```

```python
sns.countplot(
    x = "Embarked",
    data = df,
    hue = "Embarked"
)
```

b. pichart

```python
survival_counts = df["Survived"].value_counts()

print(survival_counts)

plt.pie(
    survival_counts,
    labels=["Did not survive", "Survived"],
    autopct = "%1.1f%%"
    )
plt.show()
```

```python
sex_counts = df["Sex"].value_counts()

print(sex_counts)

plt.pie(
    sex_counts,
    labels=["male", "female"],
    autopct = "%1.1f%%"
    )
plt.show()
```

```python
pclass_counts = df["Pclass"].value_counts()

print(pclass_counts)

plt.pie(
    pclass_counts,
    labels=["class-3", "class-1", "class-1"],
    autopct = "%1.1f%%"
    )
plt.show()
```

2. Numeraical Data

a. histogram

```python
plt.hist(df["Age"].dropna(), color="skyblue", edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Number of passengers")
plt.title("Age Distribution of Titanic Passengers")
plt.show()
```

b. Histogram + KDE

```python
sns.histplot(df["Age"].dropna(), bins = 10, kde = True)
plt.xlabel("Age")
plt.ylabel("Number of passengers")
plt.title("Age Distribution of Titanic Passengers")
plt.show()
```

```python
df["Age"].skew()  
```

c. boxplot

```python
# sns.boxplot(x= "Age", data = df)
sns.boxplot(x=df["Age"])
plt.title("Age Distribution (Boxplot)")
plt.show()
```
