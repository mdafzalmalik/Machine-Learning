```python
all_pages = []
```

```python
for i in range(1, 10):
    try:
        response = requests.get(
            f"https://api.themoviedb.org/3/trending/movie/day?api_key=2696d2a038461087f9e32f5c386ecc62&page={i}",
            timeout=10  # wait max 10 seconds
        )
        response.raise_for_status()  # raises error if status is not 200
        data = response.json()
        trending_movies = data.get("results", [])
        page_wise_df = pd.DataFrame(trending_movies)
        all_pages.append(page_wise_df)
    except requests.exceptions.Timeout:
        print(f"Timeout on page {i}, skipping...")
    except requests.exceptions.RequestException as e:
        print(f"Error on page {i}: {e}")

df = pd.concat(all_pages, ignore_index=True)
```

```python
df = df[["title", "release_date", "vote_average", "overview"]]

df = df.rename(columns={
    "title": "Movie Title",
    "release_date": "Release Date",
    "vote_average": "Rating",
    "overview": "Description"
})
```
