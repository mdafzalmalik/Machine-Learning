```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
all_pages = []
for j in range(1, 2):
    
    url = f"https://www.ambitionbox.com/list-of-companies?page={j}"

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9"
    }

    response = requests.get(url, headers=headers)
    webpage = response.text
    soup = BeautifulSoup(webpage, "html.parser")

    comp_details = soup.find_all("div", class_="companyCardWrapper")

    comp_name = []
    ratings = []
    reviews = []
    salaries = []
    interviews = []
    jobs = []
    benefits = []
    for i in comp_details:
        comp_name.append(i.h2.text.strip())
        ratings.append(i.find("div", class_="rating_text rating_text--md").text.strip())
        actions = i.find_all("a", class_="companyCardWrapper__ActionWrapper")
        reviews.append(actions[0].span.text)
        salaries.append(actions[1].span.text)
        interviews.append(actions[2].span.text)
        jobs.append(actions[3].span.text)
        benefits.append(actions[4].span.text)


    d = {"Company Name": comp_name, 
         "Ratings": ratings, 
         "Reviews": reviews, 
         "Salaries": salaries, 
         "Interviews": interviews, 
         "Jobs": jobs,
         "Benefits": benefits}
    temp_df = pd.DataFrame(d)
    all_pages.append(temp_df)

    time.sleep(1)
```

```python
df = pd.concat(all_pages, ignore_index=True)
```
