```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
sales_amount = np.random.randint(20, size=(5,3))
sales_amount
```

```python
# creating weekly sales 
weekly_sales = pd.DataFrame(sales_amount,
                            index=["Mon", "Tues", "Wed", "Thurs", "Fri"],
                            columns=["Almon Butter", "Peanut Butter", "Chasew Butter"])
weekly_sales
```

```python
# creating prices array
prices = np.array([10, 8, 12])
prices
```
